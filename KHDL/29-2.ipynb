```python
1. Viết chương trình có sử dụng class:
- Sinh viên chọn class bất kỳ.
- Yêu cầu có tham số nhập từ bàn phím
```

```python
import math
# z = a + bi
class Complex:
    def __init__(self, real = 0, imaginary = 0):
        self.a = real
        self.b = imaginary
    def __str__(self):
        return f"{self.a} + {self.b}i"
    def __add__(self, other):
        return Complex(self.a + other.a, self.b + other.b)
    def __sub__(self, other):
        return Complex(self.a - other.a, self.b - other.b)
    def __mul__(self, other):
        return Complex(self.a * other.a - self.b * other.b, self.a * other.b + self.b * other.a)
    def __truediv__(self, other):
        return Complex((self.a * other.a + self.b * other.b)/(other.a * other.a + other.b * other.b), (self.b * other.a - self.a * other.b)/(other.a * other.a + other.b * other.b))
    def __abs__(self):
        return math.sqrt(self.a * self.a + self.b * self.b)
    def __eq__(self, other):
        return self.a == other.a and self.b == other.b
    def input(self):
        self.a = int(input("Enter real part: "))
        self.b = int(input("Enter imaginary part: "))

z1 = Complex()
z2 = Complex()
z1.input()
z2.input()

print(z1)
print(z2)
print(z1 + z2)
print(z2 - z1)
print(z1 * z2)
print(z1 / z2)
print(abs(z1))
```

```python
2. Vẽ đồ thị hàm số f(x,y,z) (đồ thị 2d, 3d, scatter) với f(x,y,z) tùy ý.
- Yêu cầu f có dạng phân số.
- Yêu cầu f có các hàm siêu việt (hàm logarit,...)
- Trên tử và mẫu đều phải có cả tham số x, y, z.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y, z = 0):
    numerator = np.sin(x) + np.cos(y) + np.exp(z)
    denominator = x + y + z + 1
    return numerator / denominator

# Generate x and y values
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
#z = np.linspace(-5, 5, 10)

# Generate a grid of points
X, Y = np.meshgrid(x, y)

# Compute the function values
F = f(X, Y)

# Set up a figure twice as tall as it is wide
fig = plt.figure(figsize=plt.figaspect(2.))
fig.suptitle('f(x,y,z)')

# First subplot
ax = fig.add_subplot(2, 1, 1)
ax.plot(X,Y,F)
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')

# Second subplot
ax = fig.add_subplot(2, 1, 2, projection='3d')

ax.scatter(X,Y,F)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()

plt.show()
```
