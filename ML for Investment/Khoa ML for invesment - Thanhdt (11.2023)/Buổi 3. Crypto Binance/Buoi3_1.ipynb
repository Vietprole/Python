```python
import ccxt
import pandas as pd

# Khởi tạo sàn giao dịch Binance
exchange = ccxt.binance()

# Đặt cặp giao dịch (BTC/USDT) và khoảng thời gian (H1 là mỗi giờ)
symbol = 'ETH/USDT'
timeframe = '1m'

# Lấy lịch sử giá
ohlcv = exchange.fetch_ohlcv(symbol, timeframe)

ohlcv
```

```python
# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])

# Chuyển đổi timestamp sang dạng ngày tháng
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

# In ra lịch sử giá
print(df)
```

```python
import ccxt
import pandas as pd

exchange = ccxt.binance()
symbol = 'BTC/USDT'
timeframe = '1m'

# Lấy 500 bản ghi đầu tiên
ohlcv1 = exchange.fetch_ohlcv(symbol, timeframe, limit=1000)

# Lấy 500 bản ghi tiếp theo
ohlcv2 = exchange.fetch_ohlcv(symbol, timeframe, limit=1000, since=ohlcv1[-1][0] + 1)

# Kết hợp kết quả
ohlcv = ohlcv1 + ohlcv2

df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
print(df)
```
