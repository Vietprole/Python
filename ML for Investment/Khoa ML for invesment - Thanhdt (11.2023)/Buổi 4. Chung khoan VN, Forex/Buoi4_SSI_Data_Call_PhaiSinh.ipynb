```python
import pandas as pd
import time
from datetime import datetime, timedelta
from ssi_fc_data import fc_md_client, model
import configDataSSL

# Chuyển đổi ngày từ string sang datetime
def parse_date(date_string):
    return datetime.strptime(date_string, '%d/%m/%Y')

# Khởi tạo client
client = fc_md_client.MarketDataClient(configDataSSL)

# Định nghĩa ngày bắt đầu và kết thúc
from_date = '01/10/2023'
to_date = '13/01/2024'

# Chuyển đổi ngày
start_date = parse_date(from_date)
end_date = parse_date(to_date)

# Lưu trữ dữ liệu của từng khối
data_blocks = []

current_date = start_date
while current_date <= end_date:
    next_date = current_date + timedelta(days=15)
    if next_date > end_date:
        next_date = end_date

    # Lấy dữ liệu cho khối ngày hiện tại
    block_data = client.intraday_ohlc(configDataSSL, model.intraday_ohlc('VN30F1M', current_date.strftime('%d/%m/%Y'), next_date.strftime('%d/%m/%Y'), 1, 9999, True, 1))
    block_data = block_data['data']
    df = pd.DataFrame(block_data)
    df['Time'] = df['TradingDate'] + " " + df['Time']

    # Chuyển định dạng số
    df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
    df['High'] = pd.to_numeric(df['High'], errors='coerce')
    df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
    df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')

    # Thêm khối dữ liệu vào danh sách
    data_blocks.append(df)

    # Chuẩn bị cho khối ngày tiếp theo
    current_date = next_date + timedelta(days=1)
    time.sleep(10)

# Ghép tất cả các khối dữ liệu thành một DataFrame
final_df = pd.concat(data_blocks)

# Trả về DataFrame
final_df
```
