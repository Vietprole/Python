```python
!pip install distfit

from distfit import distfit
import numpy as np

# To d liu gi nh cho  tui nam v n
# V d:  tui nam theo phn b chun vi trung bnh 35 v  lch chun 10
age_male = np.random.normal(35, 10, 500)

# V d:  tui n theo phn b chun vi trung bnh 30 v  lch chun 8
age_female = np.random.normal(30, 8, 500)

# Khi to i tng distfit cho nam
dist_male = distfit()

# Fit phn b vi d liu  tui nam
dist_male.fit_transform(age_male)

# Hin th kt qu v v biu  cho nam
print(dist_male.summary)
dist_male.plot()

# Khi to i tng distfit cho n
dist_female = distfit()

# Fit phn b vi d liu  tui n
dist_female.fit_transform(age_female)

# Hin th kt qu v v biu  cho n
print(dist_female.summary)
dist_female.plot()
```
