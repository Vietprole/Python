```python
!pip install distfit

from distfit import distfit
import numpy as np

# Tạo dữ liệu giả định cho độ tuổi nam và nữ
# Ví dụ: Độ tuổi nam theo phân bố chuẩn với trung bình 35 và độ lệch chuẩn 10
age_male = np.random.normal(35, 10, 500)

# Ví dụ: Độ tuổi nữ theo phân bố chuẩn với trung bình 30 và độ lệch chuẩn 8
age_female = np.random.normal(30, 8, 500)

# Khởi tạo đối tượng distfit cho nam
dist_male = distfit()

# Fit phân bố với dữ liệu độ tuổi nam
dist_male.fit_transform(age_male)

# Hiển thị kết quả và vẽ biểu đồ cho nam
print(dist_male.summary)
dist_male.plot()

# Khởi tạo đối tượng distfit cho nữ
dist_female = distfit()

# Fit phân bố với dữ liệu độ tuổi nữ
dist_female.fit_transform(age_female)

# Hiển thị kết quả và vẽ biểu đồ cho nữ
print(dist_female.summary)
dist_female.plot()
```
