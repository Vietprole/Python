# Pyfolio VCB bang YFinance

```python
import yfinance as yf
import pyfolio as pf
import pandas as pd

# Bước 1: Tải dữ liệu cổ phiếu VCB
vcb = yf.download('VCB.VN', start='2023-01-01', end='2024-01-01')

# Bước 2: Tính toán lợi nhuận hàng ngày
# Lưu ý: Pyfolio yêu cầu dữ liệu lợi nhuận, không phải giá cổ phiếu.
vcb_returns = vcb['Close'].pct_change().dropna()

# Bước 3: Phân tích hiệu suất với Pyfolio
# Trước khi tạo tear sheet, đảm bảo chỉ mục là kiểu datetime
vcb_returns.index = pd.to_datetime(vcb_returns.index)

# Tạo một bảng phân tích đơn giản cho lợi nhuận VCB
pf.create_simple_tear_sheet(vcb_returns)
```
