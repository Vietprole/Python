```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Tạo dữ liệu giả định với thông tin mới
data = {
    'mau_vang': [1, 1, 1, 0, 0, 1, 1, 0],  # 1 la mau vang, 0 la khac mau vang: Hươu cao cổ và Hổ vằn màu vàng
    'chieu_cao': [10, 10, 10, 9, 9, 11, 11, 4],  # Hươu cao cổ và Hổ vằn có chiều cao là 10
    'dong_vat': ['huou cao co', 'huou cao co', 'huou cao co', 'voi', 'voi', 'ho van', 'ho van', 'khi']  # Nhãn 3 - 2 - 2 - 1 = 8 con
}

df = pd.DataFrame(data)

# Mã hóa nhãn từ văn bản sang số
df['dong_vat'] = df['dong_vat'].map({'huou cao co': 0, 'voi': 1, 'ho van': 2, 'khi': 3})

# Đặt các đặc trưng vào X và nhãn vào y
X = df[['mau_vang', 'chieu_cao']]
y = df['dong_vat']

# Tạo và huấn luyện mô hình cây quyết định: random_state được sử dụng để đảm bảo rằng kết quả của việc huấn luyện mô hình là nhất quán và 
# có thể tái tạo được; nghĩa là mỗi lần bạn chạy mã này với cùng một dữ liệu đầu vào, bạn sẽ nhận được cùng một mô hình cây quyết định
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X, y)

# Vẽ cây quyết định
fig, ax = plt.subplots(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=['mau_vang', 'chieu_cao'], class_names=['huou cao co', 'voi', 'ho van', 'khi'])
plt.show()
```
