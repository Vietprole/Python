```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# To d liu gi nh vi thng tin mi
data = {
    'mau_vang': [1, 1, 1, 0, 0, 1, 1, 0],  # 1 la mau vang, 0 la khac mau vang: Hu cao c v H vn mu vng
    'chieu_cao': [10, 10, 10, 9, 9, 11, 11, 4],  # Hu cao c v H vn c chiu cao l 10
    'dong_vat': ['huou cao co', 'huou cao co', 'huou cao co', 'voi', 'voi', 'ho van', 'ho van', 'khi']  # Nhn 3 - 2 - 2 - 1 = 8 con
}

df = pd.DataFrame(data)

# M ha nhn t vn bn sang s
df['dong_vat'] = df['dong_vat'].map({'huou cao co': 0, 'voi': 1, 'ho van': 2, 'khi': 3})

# t cc c trng vo X v nhn vo y
X = df[['mau_vang', 'chieu_cao']]
y = df['dong_vat']

# To v hun luyn m hnh cy quyt nh: random_state c s dng  m bo rng kt qu ca vic hun luyn m hnh l nht qun v 
# c th ti to c; ngha l mi ln bn chy m ny vi cng mt d liu u vo, bn s nhn c cng mt m hnh cy quyt nh
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X, y)

# V cy quyt nh
fig, ax = plt.subplots(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=['mau_vang', 'chieu_cao'], class_names=['huou cao co', 'voi', 'ho van', 'khi'])
plt.show()
```
