# Tao nen Doji tu du lieu fix

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

def draw_candlestick(ax, open_price, high_price, low_price, close_price, x_position):
    # Vẽ đường High-Low
    high_low_line = mlines.Line2D([x_position, x_position], [low_price, high_price], color='black')
    ax.add_line(high_low_line)
    
    # Xác định màu sắc dựa trên giá Open và Close
    color = 'green' if open_price <= close_price else 'red'
    # Vẽ hình chữ nhật Open-Close
    rect_height = max(close_price - open_price, 0.01)  # Tránh trường hợp không có chiều cao
    open_close_rect = plt.Rectangle((x_position - 0.1, open_price), 0.2, rect_height, 
                                    facecolor=color, edgecolor='black')
    ax.add_patch(open_close_rect)
    
    # Thêm nhãn giá
    ax.text(x_position, high_price, f'High: {high_price}', verticalalignment='bottom', horizontalalignment='center', fontsize=8)
    ax.text(x_position, low_price, f'Low: {low_price}', verticalalignment='top', horizontalalignment='center', fontsize=8)
    ax.text(x_position-0.1, open_price, f'Open: {open_price}', verticalalignment='center', horizontalalignment='right', fontsize=8)
    ax.text(x_position+0.1, close_price, f'Close: {close_price}', verticalalignment='center', horizontalalignment='left', fontsize=8)

# Khởi tạo plot
fig, ax = plt.subplots(figsize=(10, 6))

# Vẽ các candlestick
draw_candlestick(ax, 10, 15, 5, 11, 1)  # Long Legged
draw_candlestick(ax, 20, 25, 15, 15.1, 2)  # Gravestone
draw_candlestick(ax, 30, 35, 28, 34.9, 3)  # Dragonfly
draw_candlestick(ax, 40, 45, 35, 40, 4)  # Star

# Thiết lập giới hạn cho biểu đồ
ax.set_xlim(0, 5)
ax.set_ylim(0, 50)

# Thiết lập nhãn (tùy chọn)
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['LONG LEGGED', 'GRAVETONE', 'DRAGONFLY', 'STAR'])
ax.set_ylabel('Price')

# Hiển thị plot
plt.show()
```
